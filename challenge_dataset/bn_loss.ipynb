```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import wandb
```

```python
api = wandb.Api()

runs = api.runs("jschuetzke/synthetic-benchmark")
history = {}
for run in runs:
    if run.config["evaluation"] == "challenge" or  run.config["evaluation"] == "modifications":
        # run.name is the name of the run
        history[run.name] = run.history()     
```

```python
plt.rcParams.update({'font.size': 14})
plt.rcParams.update({'lines.linewidth': 3})
```

```python
fig,ax = plt.subplots(3,2,figsize=(12,10), sharey=True)

ax[0,0].plot(history["cnn_bn-4-noact"]['val_loss'].values)
ax[0,0].plot(history["cnn_bn-3-noact"]['val_loss'].values)
ax[0,0].plot(history["cnn_bn-2-noact"]['val_loss'].values)
ax[0,0].plot(history["cnn_bn-1-noact"]['val_loss'].values)
ax[0,0].plot(history["cnn_bn-0-noact"]['val_loss'].values)
ax[0,1].plot(history["cnnbn_mod-4"]['val_loss'].values)
ax[0,1].plot(history["cnnbn_mod-3"]['val_loss'].values)
ax[0,1].plot(history["cnnbn_mod-2"]['val_loss'].values)
ax[0,1].plot(history["cnnbn_mod-1"]['val_loss'].values)
ax[0,1].plot(history["cnnbn_mod-0"]['val_loss'].values)
ax[0,1].set_ylim(0.3, 5)
ax[1,0].plot(history["resnet-4"]['val_loss'].values)
ax[1,0].plot(history["resnet-3"]['val_loss'].values)
ax[1,0].plot(history["resnet-2"]['val_loss'].values)
ax[1,0].plot(history["resnet-1"]['val_loss'].values)
ax[1,0].plot(history["resnet-0"]['val_loss'].values)
ax[1,1].plot(history["resnet_mod-4"]['val_loss'].values)
ax[1,1].plot(history["resnet_mod-3"]['val_loss'].values)
ax[1,1].plot(history["resnet_mod-2"]['val_loss'].values)
ax[1,1].plot(history["resnet_mod-1"]['val_loss'].values)
ax[1,1].plot(history["resnet_mod-0"]['val_loss'].values)
ax[1,1].set_ylim(0.3, 5)
ax[2,0].plot(history["cnn6_bn-4"]['val_loss'].values)
ax[2,0].plot(history["cnn6_bn-3"]['val_loss'].values)
ax[2,0].plot(history["cnn6_bn-2"]['val_loss'].values)
ax[2,0].plot(history["cnn6_bn-1"]['val_loss'].values)
ax[2,0].plot(history["cnn6_bn-0"]['val_loss'].values)
ax[2,1].plot(history["cnn6-4-relu"]['val_loss'].values)
ax[2,1].plot(history["cnn6-3-relu"]['val_loss'].values)
ax[2,1].plot(history["cnn6-2-relu"]['val_loss'].values)
ax[2,1].plot(history["cnn6-1-relu"]['val_loss'].values)
ax[2,1].plot(history["cnn6-0-relu"]['val_loss'].values)
ax[2,0].set_ylim(0.3, 5)
#fig.suptitle("Evaluation of Batch Normalization")
ax[0,0].set_title("Networks with Batch Normalization")
ax[0,1].set_title("Networks without Batch Normalization")
#ax[0,0].set_ylabel("Model CNN BN")
#ax[1,0].set_ylabel("Model Resnet")
#ax[2,0].set_ylabel("Model CNN6")
fig.supylabel("Validation Loss")
fig.supxlabel("Epoch")

fig.patches.extend([plt.Rectangle((0.44,0.87),0.17,0.07, edgecolor='black',
                                  fill=True, fc='white', alpha=1., zorder=3,
                                  transform=fig.transFigure, figure=fig),
                    plt.Rectangle((0.44,0.48),0.17,0.07, edgecolor='black',
                                  fill=True, fc='white', alpha=1., zorder=3,
                                  transform=fig.transFigure, figure=fig),
                    plt.Rectangle((0.44,0.263),0.17,0.07, edgecolor='black',
                                  fill=True, fc='white', alpha=1., zorder=3,
                                  transform=fig.transFigure, figure=fig)])
fig.text(0.485,0.896,"CNN BN", fontsize=18)
fig.text(0.49,0.508,"Resnet", fontsize=18)
fig.text(0.5,0.289,"CNN6", fontsize=18)
plt.tight_layout()
plt.show()
#plt.savefig("batch_normalization_loss.eps", bbox_inches='tight')
```
